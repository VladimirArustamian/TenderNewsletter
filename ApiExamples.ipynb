```python
import json
import requests
from datetime import datetime
from google.oauth2 import service_account
from google.auth.transport.requests import Request

# Refresh the credentials to obtain the ID token
credentials.refresh(Request())
id_token = credentials.token

# Set up headers with the ID token
headers = {
    'Authorization': f'Bearer {id_token}',
    'Content-Type': 'application/json'
}

# Example search parameters
search_params = {
    "query": "construction project",
    "match_count": 30, #maxiumum 30
    "columns": ["content", "title", "description"],
    "notice_langs_param": ["ENG", "FRA"], # Use that field
    "tender_deadline_min": datetime(2023, 1, 1).isoformat(),
    "issue_date_min": datetime(2022, 6, 1).isoformat(),
    "issue_date_max": datetime(2023, 12, 31).isoformat()
}

# Optional search ID
# search_id = "123e4567-e89b-12d3-a456-426614174000"

# Prepare the payload
payload = {
    "query": search_params["query"],
    "match_count": search_params.get("match_count", 30),
    "columns": search_params.get("columns", ["content"]),
    "notice_langs_param": search_params.get("notice_langs_param", ["ENG"]),
    "tender_deadline_min": search_params.get("tender_deadline_min"),
    "issue_date_min": search_params.get("issue_date_min"),
    "issue_date_max": search_params.get("issue_date_max")
}

# If a search_id is provided, include it as a query parameter
params = {}
if search_id:
    params["searchId"] = search_id

try:
    # Make the POST request to the search API with headers
    response = requests.post(SEARCH_API_URL, headers=headers, json=payload, params=params)

    # Handle the response
    if response.status_code == 200:
        data = response.json()
        print("Search Successful:")
        print(json.dumps(data, indent=2))
    elif response.status_code == 204:
        print("No Content found for the given query.")
    else:
        print(f"Error: Received status code {response.status_code}")
        print("Response:", response.text)
except requests.exceptions.RequestException as e:
    print(f"An error occurred while making the request: {e}")
```
